```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Preparing the DataFrame

### Qn #1 Compare  countries by their health condition

```python
plt.figure(figsize=(20,5))
sns.barplot(x= df["country"].unique(),y= df["health"])
plt.title("Ans: Hong Kong got the overall rank of 100 regarding their countries health condition")
plt.xticks(rotation=90,size = 7)
plt.show()
```

### Qn #2 Which country has the max amount of popularity

```python
plt.figure(figsize=(25,5))
sns.barplot(x = df["country"].unique(),y = df["popularity"])
plt.title("Ans: China")
plt.xticks(rotation = 90, fontsize = 10)
plt.show()
```
